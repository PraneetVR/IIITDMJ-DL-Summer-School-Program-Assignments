```python
model = VGG16(weights='imagenet',include_top=False) #loaading VGG16

for layer in model.layers:
    layer.trainable=False     #freezing layers

trainable_params=sum([layer.count_params() for layer in model.layers if layer.trainable])
non_trainable_params=sum([layer.count_params() for layer in model.layers if not layer.trainable])

print("Trainable parameters:",trainable_params)
print("Non-trainable parameters:",non_trainable_params)   #printing trainable and non trainable parameters
```
