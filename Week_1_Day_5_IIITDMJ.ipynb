#Day 5 Assignment

[Implement a single-layer neural network with:

Input: shape (1, 3)

Weights: shape (3, 1) (initialize manually)

Bias: scalar or shape (1,)

Activation: Sigmoid

Loss: Binary Cross-Entropy

Ground truth label: y_true = 1.0]

```python
import tensorflow as tf
import torch
import numpy as np
```

#Task 1
(Implement a single-layer neural network with TensorFlow (Keras backend))

```python
x = tf.constant([[1.0, 3.0]])
w = tf.Variable([[3.0], [1.0]])
b = tf.Variable([1.0])
y_true = tf.constant([1.0])

z = tf.matmul(x, w) + b

a = tf.math.sigmoid(z)
print("TensorFlow Output(2 inputs + bias + sigmoid):", a.numpy())
print()

loss = - (y_true * tf.math.log(a) +
         (1 - y_true) * tf.math.log(1 - a))
print("loss :",loss.numpy())
```

#Task 2
(Implement a single-layer neural network with Pytorch)

```python
x = torch.tensor([[1.0, 3.0]])
w = torch.nn.Parameter(torch.tensor([[3.0], [1.0]]))
b = torch.nn.Parameter(torch.tensor([1.0]))
y_true = torch.tensor([[1.0]])

z = torch.mm(x, w) + b

a = torch.sigmoid(z)
print("Torch Output (2 inputs + bias + sigmoid):",a.detach().numpy())
print()

loss = - (y_true * torch.log(a) + (1 - y_true) * torch.log(1 - a))

loss.backward()

print("loss :",loss.detach().numpy())
print()

print("Gradients of weights:",w.grad)
print("Gradient of bias (b):",b.grad)
```
