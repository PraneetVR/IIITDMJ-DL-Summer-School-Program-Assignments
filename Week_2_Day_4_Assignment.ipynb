#Q1. Load a Transformer Model and Tokenizer

```python
from transformers import AutoModelForSequenceClassification, AutoTokenizer, T5ForConditionalGeneration

model_name = "bert-base-uncased"
model = AutoModelForSequenceClassification.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)

print(f"Model Name: {model_name}")

# Calculate the total number of parameters
total_params = sum(p.numel() for p in model.parameters())
print(f"Total Parameters: {total_params}")

model_name = "t5-small"
model = T5ForConditionalGeneration.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)

print(f"Model Name: {model_name}")

# Calculate the total number of parameters
total_params = sum(p.numel() for p in model.parameters())
print(f"Total Parameters: {total_params}")

model_name = "Helsinki-NLP/opus-mt-en-hi"
model = T5ForConditionalGeneration.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)

print(f"Model Name: {model_name}")

# Calculate the total number of parameters
total_params = sum(p.numel() for p in model.parameters())
print(f"Total Parameters: {total_params}")
```

#Q2. Simple Summarization using T5

```python
text = "Natural Language Processing is a part of Artificial Intelligence that focuses on interaction between computers and human language. It enables applications like chatbots, translation, and summarization."

input_text = "summarize: " + text

input_ids = tokenizer.encode(input_text, return_tensors="pt")

summary_ids = model.generate(input_ids, max_length=50, min_length=10, num_beams=4, early_stopping=True)

summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)

print("Original Text:",text)
print()
print("\nSummarized Output:", summary)
```

#Q4. Tokenize and Decode a Sentence using BERT Tokenizer

```python
tokenizer_bert = AutoTokenizer.from_pretrained("bert-base-uncased")

sentence = "Transformers make NLP tasks easier."

tokenized_sentence = tokenizer_bert(sentence, return_tensors="pt")

token_ids = tokenized_sentence["input_ids"][0].tolist()
print("Token IDs:", token_ids)

decoded_tokens = tokenizer_bert.convert_ids_to_tokens(token_ids)
print("Decoded Tokens:", decoded_tokens)
```
