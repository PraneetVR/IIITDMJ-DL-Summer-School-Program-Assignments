```python
MD=np.array([[4,2],[1,3]])
print("Matrix D = \n",MD)
print()
eigenvalues, eigenvectors = eig(MD)

print("Eigenvalues:", eigenvalues)
print("Eigenvectors:\n", eigenvectors)
print()
origin = np.zeros((2, 2))
plt.quiver(*origin, eigenvectors[0, :], eigenvectors[1, :],scale=5)
plt.grid(True)
plt.show()
```

#P2
(Symbolic Differentiation and Gradient Step )

1.

```python
x = sp.symbols('x')
f = 3*x**3 - 5*x**2 + 2*x - 1
df = sp.diff(f, x)
print("f'(x) =", df)
print()
#derivative at x = 2
df_at_x = df.subs(x, 2)
print("f'(2) =", df_at_x)
```

2.

```python
y = sp.symbols('y')
f2 = x**2 + y**2
#gradient
df_dx = sp.diff(f2, x)
df_dy = sp.diff(f2, y)
print("f'(x) =", df_dx)
print("f'(y) =", df_dy)
print()
#Perform one gradient descent step from (x=3, y=4) using learning rate = 0.1
learning_rate = 0.1
x_new = 3 - learning_rate * df_dx.subs([(x, 3), (y, 4)])
y_new = 4 - learning_rate * df_dy.subs([(x, 3), (y, 4)])
print("New values of x and y:", (x_new, y_new))
```

(Probability with Dice and Coin Simulation)

1.

```python
die_rolls=np.random.randint(1, 7, size=1000)
print(die_rolls)
plt.hist(die_rolls,edgecolor='black')
plt.title('Histogram of 1000 Fair Die Rolls')
plt.show()
print()
mean=np.mean(die_rolls)
variance=np.var(die_rolls)
print(f"Empirical mean: {mean:.2f}")
print(f"Empirical variance: {variance:.2f}")
```

2.

```python
coin_outcomes = np.random.choice(['Heads', 'Tails'], size=500, p=[0.7, 0.3])
unique, counts = np.unique(coin_outcomes, return_counts=True)
outcome_counts = dict(zip(unique, counts))
print(f"Counts: {outcome_counts}")
print()
#bar graph
plt.bar(outcome_counts.keys(), outcome_counts.values())
plt.title('Biased Coin Tosses (500 trials, P(heads)=0.7)')
plt.show()
```
