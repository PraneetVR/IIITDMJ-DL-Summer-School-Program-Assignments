```python
from sklearn.datasets import load_iris
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
db = load_iris()
db
```

```python
x=db.data
y=db.target
x_train,x_test,y_train,y_test=train_test_split(x, y, test_size=0.2, random_state=42)
```

###logistic regression

```python
LogReg=LogisticRegression().fit(x_train, y_train)
y_predl=LogReg.predict(x_test)
acc=accuracy_score(y_test, y_predl)
cm=confusion_matrix(y_test, y_predl)
print(f"Accuracy: {acc}")
print()
print(f"Confusion Matrix: \n{cm}")
```

```python
sns.heatmap(cm, annot=True)
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```

###decision tree

```python
DT = DecisionTreeClassifier(random_state=42).fit(x_train, y_train)
y_predD = DT.predict(x_test)

print("Accuracy:", accuracy_score(y_test, y_predD))
print()
print("Confusion Matrix:\n", confusion_matrix(y_test, y_predD))
```

###Neural Net

```python
NN=MLPClassifier(hidden_layer_sizes=(1),max_iter=3000,random_state=42).fit(x_train, y_train)
y_predNN = NN.predict(x_test)

print("Accuracy:", accuracy_score(y_test, y_predNN))
print()
print("Confusion Matrix:\n", confusion_matrix(y_test, y_predNN))
```

Neural networks can model more complex, non-linear relationships in data due to their layered structure and non-linear activation functions.

Unlike logistic regression (which is linear) and decision trees (which split data based on simple rules), a neural net with even one hidden layer can approximate more intricate boundaries between classes.

Neural nets are less interpretable than decision trees and logistic regression, but they can generalize better on complex datasets, especially as the data or feature space grows.

On simple, well-separated data like Iris, all three methods may perform similarly, but neural nets show their advantage on more complex or high-dimensional datasets.

```python
X=np.array([[0,0],[0,1],[1,0],[1,1]])
y=np.array([0,0,0,1])
```

```python
class Perceptron:
    def __init__(self,learning_rate=0.1,epochs=20,activation='step'):
        self.lr=learning_rate
        self.epochs=epochs
        self.activation=activation
        self.weights=None
        self.bias=None

    def _step(self,x):
        return np.where(x >= 0,1,0)

    def _relu(self,x):
        return np.maximum(0,x)

    def _sigmoid(self,x):
        return 1/(1 + np.exp(-x))

    def _activate(self,x):
        if self.activation=='step':
            return self._step(x)
        elif self.activation=='relu':
            return self._relu(x)
        elif self.activation=='sigmoid':
            return self._sigmoid(x)
        else:
            raise ValueError("Unknown activation")

    def fit(self,X,y):
        n_samples,n_features=X.shape
        self.weights=np.zeros(n_features)
        self.bias=0

        for _ in range(self.epochs):
            for idx,x_i in enumerate(X):
                linear_output=np.dot(x_i,self.weights) + self.bias
                y_pred=self._step(linear_output)
                update=self.lr*(y[idx] - y_pred)
                self.weights += update*x_i
                self.bias += update

    def predict(self,X):
        linear_output=np.dot(X,self.weights) + self.bias
        return self._activate(linear_output)
```

```python
activations = ['step','relu','sigmoid']
titles = ['Perceptron (Step)','Perceptron (ReLU)','Perceptron (Sigmoid)']

for i,act in enumerate(activations):
    p=Perceptron(learning_rate=0.1,epochs=20,activation=act)
    p.fit(X,y)
    y_pred=p.predict(X)

    xx,yy=np.meshgrid(np.linspace(-0.5, 1.5, 200),np.linspace(-0.5, 1.5, 200))
    grid=np.c_[xx.ravel(),yy.ravel()]
    zz=p.predict(grid).reshape(xx.shape)

    plt.subplot(1,3,i+1)
    plt.contourf(xx,yy,zz,levels=50)
    plt.scatter(X[:, 0],X[:, 1],edgecolors='k')
    plt.title(titles[i])
    plt.xlabel('x1')
    plt.ylabel('x2')

plt.tight_layout()
plt.show()
```

```python
np.random.seed(42)
```

```python
x=np.linspace(0, 10, 100)
noise=np.random.normal(0,2,size=x.shape)
y=2.5*x+np.sin(x)+noise

plt.scatter(x,y)
plt.title("Synthetic Regression Data")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)

x_train = x_train.reshape(-1, 1)
x_test = x_test.reshape(-1, 1)
```

```python
def build_model():
    model=tf.keras.Sequential([
        tf.keras.layers.Dense(32,activation='relu',input_shape=(1,)),
        tf.keras.layers.Dense(16,activation='relu'),
        tf.keras.layers.Dense(1)
    ])
    return model
```

```python
optimizers={'SGD': tf.keras.optimizers.SGD(),'Adam': tf.keras.optimizers.Adam(),'RMSProp': tf.keras.optimizers.RMSprop()}

histories={}
rmses={}

for name,opt in optimizers.items():
    model=build_model()
    model.compile(optimizer=opt,loss='mse')
    history=model.fit(x_train,y_train,epochs=100,verbose=0,validation_split=0.1)
    histories[name]=history
    y_pred=model.predict(x_test).flatten()
    rmse=np.sqrt(mean_squared_error(y_test,y_pred))
    rmses[name]=rmse
    print(f"{name} RMSE: {rmse}")

    plt.plot(history.history['loss'], label=f'{name} Train')
    plt.plot(history.history['val_loss'], label=f'{name} Val')
    plt.title(f'Loss Curve ({name})')
    plt.xlabel('Epoch')
    plt.ylabel('MSE Loss')
    plt.legend()
    plt.show()
```

Adam is generally the best optimizer for small neural nets on regression tasks due to its adaptive learning rate and momentum, leading to faster and more reliable convergence compared to SGD and RMSProp.

```python
from tensorflow.keras import layers, models
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input
from tensorflow.keras.datasets import mnist
```

```python
(x_train,y_train),(x_test,y_test)=mnist.load_data()
x_train=x_train[...,None]/255.0
x_test=x_test[...,None]/255.0

model = models.Sequential([
    layers.Input(shape=(28, 28, 1)),
    layers.Conv2D(32, 3, activation='relu'),
    layers.MaxPooling2D(),
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax')
])

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.fit(x_train, y_train, epochs=3, batch_size=128, verbose=2)
test_loss, test_acc = model.evaluate(x_test, y_test, verbose=0)

print("Keras Model Summary:")
model.summary()
print(f"Keras Test Accuracy: {test_acc:.4f}")
```

```python
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader
from torchvision import datasets, transforms

transform = transforms.Compose([transforms.ToTensor()])
train_set = datasets.MNIST(root='./data', train=True, download=True, transform=transform)
test_set = datasets.MNIST(root='./data', train=False, download=True, transform=transform)
train_loader = DataLoader(train_set, batch_size=128, shuffle=True)
test_loader = DataLoader(test_set, batch_size=128, shuffle=False)

class SmallCNN(nn.Module):
    def __init__(self):
        super().__init__()
        self.conv = nn.Conv2d(1, 32, 3)
        self.pool = nn.MaxPool2d(2)
        self.fc1 = nn.Linear(32*13*13, 64)
        self.fc2 = nn.Linear(64, 10)
    def forward(self, x):
        x = torch.relu(self.conv(x))
        x = self.pool(x)
        x = x.view(x.size(0), -1)
        x = torch.relu(self.fc1(x))
        return self.fc2(x)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = SmallCNN().to(device)
optimizer = optim.Adam(model.parameters())
criterion = nn.CrossEntropyLoss()

model.train()
for epoch in range(3):
    for images, labels in train_loader:
        images, labels = images.to(device), labels.to(device)
        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

model.eval()
correct, total = 0, 0
with torch.no_grad():
    for images, labels in test_loader:
        images, labels = images.to(device), labels.to(device)
        outputs = model(images)
        _, preds = torch.max(outputs, 1)
        correct += (preds == labels).sum().item()
        total += labels.size(0)
test_acc = correct / total

print("PyTorch Model Summary:")
print(model)
print(f"PyTorch Test Accuracy: {test_acc:.4f}")
```
