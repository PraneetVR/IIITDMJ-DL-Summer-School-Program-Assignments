#P1
(Write a NumPy function softmax(x) that takes a 1D array x and returns its softmax probabilities.)


```python
def softmax(x):
  a=np.exp(x)/np.sum(np.exp(x))
  return a

x = np.array([1.0, 2.0, 3.0])
print(softmax(x))
```

#P2
(Given these examples, classify each as AI, ML, or DL by writing print() statements.)

```python
examples = ["Self-driving car using CNNs",
            "Spam detection using Naive Bayes",
            "Expert system for medical diagnosis",
            "Face recognition using Siamese Networks",
            "Weather forecasting using linear regression"]

for i in examples:
  if "CNNs" in i:
    print(i,"is DL")
  elif "Naive Bayes" in i:
    print(i,"is ML")
  elif "linear regression" in i:
    print(i,"is ML")
  elif "Siamese Networks" in i:
    print(i,"is DL")
  else:
    print(i,"is AI")

```

#P3
(Generate a 5x5 random matrix with values between 0 and 1 using PyTorch. Compute the mean, standard deviation, and sum of all elements using PyTorch operations. Then, convert the tensor to a NumPy array and recompute these statistics using NumPy.
)

```python
#5x5 array
MV=torch.randn(5,5)
print(MV)
print()
#mean of matrix
print("mean of pytorch matrix is",torch.mean(MV))
print()
#standard deviation
print("standard deviation of pytorch matrix is",torch.std(MV))
print()
#sum
print("Sum of all elements in a pytorch matrix is",torch.sum(MV))
```

```python
#convert to numpy array
MV1=np.array(MV)
print(MV1)
print()
#mean of matrix
print("mean of the matrix is",np.mean(MV1))
print()
#standard deviation
print("standard deviation of pytorch matrix is",np.std(MV1))
print()
#sum
print("Sum of all elements in a pytorch matrix is",np.sum(MV1))
```
